```python
import pandas as pd
pd.set_option("display.max_columns", 101)
pd.set_option("display.max_rows", 101)
```

```python
data=pd.read_csv('bcc.csv')
data['radiation- first-Event Date']=pd.to_datetime(data['radiation- first-Event Date'])
data['radiation-last-Event Date']=pd.to_datetime(data['radiation-last-Event Date'])

data['Reference Event-Date of Documentation']=pd.to_datetime(data['Reference Event-Date of Documentation'])
```

```python
dates=data[['Patient ID','Reference Event-Date of Documentation','radiation- first-Event Date','radiation-last-Event Date']]
```
