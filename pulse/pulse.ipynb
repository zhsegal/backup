```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('pulse.csv')

data.columns=[name.split('-')[1] if name.startswith('Ref')else name for name in data.columns]
data.columns=[name.replace('Department ','') for name in data.columns]
data['Admission Date']=pd.to_datetime(data['Admission Date'])
```

```python
data[:30]
```

```python
count=data.groupby(pd.Grouper(key='Admission Date', freq='1Y')).count()
count
```

```python
plt.plot(count.index, count['Patient ID'])
plt.title('Solu Medrol adminstration in internal wards by year')
plt.show()
```
