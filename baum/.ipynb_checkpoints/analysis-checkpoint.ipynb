```python
import pandas as pd
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
import matplotlib.pyplot as plt
import numpy as np
pd.options.mode.chained_assignment = None 
import seaborn as sns
from scipy import stats
from scipy.stats import chi2_contingency
```

```python
data=pd.read_excel('data.xlsx')
data.set_index('ID',drop =True, inplace=True)
```

```python
data.head()
```

```python
duplicate = data[data.Full_name.duplicated(keep=False)]
duplicate
```

```python
relapse_cols=[col for col in data.columns if 'Relapse' in str(col)]
relapse=data[relapse_cols]
```

```python
rename_dict={
    'Unnamed: 246':'eos1',
    'Unnamed: 247':'eos2',
    'Unnamed: 248':'eos3',
       'Unnamed: 249':'eos4',
    'Unnamed: 250':'eos5'
}
```

```python
data.rename(columns=rename_dict, inplace=True)
```

```python
demog=['Gender','Diagnosis_Age']
comorbid=['BG_Illness_DM','BG_Illness_Neurological', 'BG_Illness_A_I', 'BG_Ilness_Neoplasm']
drugs=['Drugs_ACE', 'Drugs_ARB', 'Drugs_B_BLOCKERS', 'Drugs_DIURETICS',
       'Drugs_STATINS', 'Drugs_INSULIN', 'Drugs_ANTI_DEPR','Local_Steroid_Treatment']
sympt=['Diagnosis_Skin_Pruritus',
       'Diagnosis_Skin_Urticaria', 'Diagnosis_Skin_Blisters',
       'Diagnosis_Mocous_Oral', 'Diagnosis_Mocous_Genital',
       'Diagnosis_BSA']
labs=['IGE',       'Eosinophils_Percentage', 'Eosinophils_Absolute','peripheral_Eosinophilia']
eos_cols=['eos1', 'eos2', 'eos3', 'eos4', 'eos5']
response_cols=['FU_1_PR', 'FU_1_CR', 'FU_1_No_response','FU_3_PR', 'FU_3_CR', 'FU_3_No_response','FU_6_PR', 'FU_6_CR',
       'FU_6_No_response','FU_1_P_dosage']
no_response=['FU_1_No_response','FU_3_No_response','FU_6_No_response']
partial_response=['FU_1_PR', 'FU_3_PR','FU_6_PR']
complete_response=['FU_1_CR', 'FU_3_CR','FU_6_CR']
y_col='FU_0_No_response'
```

```python
threshold=70
eos_thresh=0

cols=demog+comorbid+drugs+sympt+labs+eos_cols+response_cols+[y_col]
final=data[cols]
final['severity'] = np.where(final['Diagnosis_BSA']<= threshold, 0, 1)

mean=final[eos_cols].mean(axis=1)
median=final[eos_cols].median(axis=1)
mini=final[eos_cols].min(axis=1)
maxi=final[eos_cols].max(axis=1)

final.loc[:,'eos_mean']=mean
final.loc[:,'eos_median']=median
final.loc[:,'eos_min']=mini
final.loc[:,'eos_max']=maxi

final=final.loc[final.eos_min>= eos_thresh]
```

```python
eos_cutoff=12

final['eos_bool'] = np.where(final['eos_mean']> eos_cutoff, 2, 1)
final['eos_bool'].loc[final.eos_max==0]=0
```

```python
dem_cols=["Gender", "Diagnosis_Age","eos_bool",'Diagnosis_Skin_Blisters','BG_Illness_DM','BG_Ilness_Neoplasm','BG_Illness_Neurological','peripheral_Eosinophilia']
```

```python
for col in dem_cols:
    contigency= pd.crosstab(final[col], final['eos_bool']) 
    c, p, dof, expected = chi2_contingency(contigency) 
    print (f'{col}: pval is {p.round(2)}')
```

```python
for col in dem_cols:
    a=final.loc[final.eos_bool==0]
    b=final.loc[final.eos_bool==1]
    c=final.loc[final.eos_bool==2]
    fvalue, pvalue = stats.f_oneway(a[col], b[col],c[col])

    print (f'{col}: pval is {pvalue}')


```

```python
desc=final[dem_cols].groupby("eos_bool").mean()
desc.columns=['percent females','mean_age', 'percent with blisters','BG_Illness_DM','BG_Ilness_Neoplasm','BG_Illness_Neurological','peripheral_Eosinophilia']
desc.index=['0 eosinophols',f'1-{eos_cutoff} eosinophols',f'{eos_cutoff}+ eosinophols']
desc.T
```

```python
desc=final[["Gender", "Diagnosis_Age","eos_bool",'Diagnosis_Skin_Blisters']].groupby("eos_bool").std()
desc.columns=['percent females','std_age', 'percent with blisters']
desc.index=['0 eosinophols',f'1-{eos_cutoff} eosinophols',f'{eos_cutoff}+ eosinophols']
desc
```

```python
g=sns.boxplot(data=final, x='eos_bool', y='Diagnosis_BSA')
g.set_xticklabels (['0',f'1-{eos_cutoff}',f'{eos_cutoff}+'])
g.set(xlabel='Mean Eosinophilc ', ylabel='BSA',title='BSA by eosinophil group')
plt.show()
```

continous eosniphils vs. reponse proportion

```python
for var in response_cols:
    group0=final.loc[final[var]==0].eos_mean
    group1=final.loc[final[var]==1].eos_mean
    
    pval=stats.ttest_ind(group0,group1)[1]
    print (f'{var}: pval is {pval}')
    pval
```

```python
pred_cut=20
final['pred_bool']=np.where(final['FU_1_P_dosage']>= pred_cut,1,0 )

pred_l=final.loc[final.pred_bool==0]
pred_h=final.loc[final.pred_bool==1]
```

```python
for var in response_cols:
    group0=pred_l.loc[pred_l[var]==0].eos_mean
    group1=pred_l.loc[pred_l[var]==1].eos_mean
    
    pval=stats.ttest_ind(group0,group1)[1]
    print (f'{var}: pval is {pval}')
    pval
```

```python
for var in response_cols:
    group0=pred_h.loc[pred_h[var]==0].eos_mean
    group1=pred_h.loc[pred_h[var]==1].eos_mean
    
    pval=stats.ttest_ind(group0,group1)[1]
    print (f'{var}: pval is {pval}')
    pval
```

BY BSA ANALYSIS

```python
eos_types=['eos_mean','eos_median','eos_min','eos_max']
print (f'number over threhols: {final.severity.sum()}')

for eosi in eos_types:
    g=sns.boxplot(data=final, x='severity', y=eosi)
    g.set_xticklabels (['Under Threshold','Over Threshold'])
    g.set(xlabel='Severity', ylabel=f'{eosi} Levels',title=f'{eosi} levels be severity, severity threshold={threshold}, eos threshold={eos_thresh}')
    
    pval=stats.ttest_ind(final[final['severity']==0][eosi],final[final['severity']==1][eosi])[1].round(3)

    
    x1, x2 = 0, 1  
    y, h, col = final[eosi].max() + 0.2, 0.2, 'k'
    plt.plot([x1, x1, x2, x2], [y, y+h, y+h, y], lw=1.5, c=col)
    plt.text((x1+x2)*.5, y+h, f"pval={pval}", ha='center', va='bottom', color=col)
    
    plt.savefig(f"figures/{eosi} by severity.png")
    plt.show()
```
