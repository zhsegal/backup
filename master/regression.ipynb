```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import matplotlib.pyplot as plt
import numpy as np
from sklearn import preprocessing
from sklearn import datasets, linear_model
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.formula.api as smf
```

```python
crea=pd.read_csv('crea.csv')
crea.set_index('id', drop=True, inplace=True)
ages=pd.read_csv('ages.csv')
ages.set_index('id', drop=True, inplace=True)
crea.dropna(thresh=10, axis=0, inplace=True)
ages.dropna(thresh=10, axis=0, inplace=True)
```

```python
crea.head()
```

```python
crea_int=crea.interpolate(method='linear', axis=1)
```

```python
crea_red=pd.DataFrame(index=crea.index)
intervals=[0,6,12,24, 60,crea.shape[1]]
df=crea_int

for i in range(len(intervals)-1):
    start=intervals[i]
    end=intervals[i+1]
    df_cut=df.iloc[:, start:end]
    means=df_cut.mean(axis=1)
    crea_red[str(end/12)]=means
```

```python
meani=crea_red.values.mean()
stdi=crea_red.values.std()
crea_red=(crea_red-meani)/stdi
crea_red.head()
```

```python
crea_red['age'] = np.random.randint(50, 70, crea_red.shape[0])
crea_red['age'] = (crea_red.age-crea_red.age.mean())/crea_red.age.std()
crea_red.columns=['half','year','two','five','all','age']
```
