```python

def randomword(length):
   letters = string.ascii_lowercase
   return ''.join(random.choice(letters) for i in range(length))
```

```python
data=pd.DataFrame(columns=[f'var_{num}' for num in range (500)])
data['id']=[randomword(7) for i in range (100000)]
data.set_index('id', drop=True, inplace=True)
```

```python
tab_size=data.shape[0]

for col_num in range(tab_size):
    mu=random.randint(0, 100)	
    std=random.randint(1, 14)	
    data.iloc[:,col_num]=s = np.random.normal(mu, std, tab_size)
```

```python
data.to_csv('sim_data.csv')
```
