```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
maindata=pd.read_csv('deps.csv')
```

```python
maindata.columns=['ID','age','date','dep','duration']
maindata.date=pd.to_datetime(maindata.date)
```

CONUT GRPHS

```python
data=maindata.loc[maindata.date>'2012-01-01']

results=pd.DataFrame()

for dep in data.dep.unique():
    temp=data.loc[data.dep==dep]
    aa=temp.groupby(pd.Grouper(freq='M', key='date'))['dep'].count().reset_index()
    aa.columns=['date','count']
    aa['dep']=dep
    results=results.append(aa)
plt.figure(figsize=(10,10))



for dep in results.dep.unique():
    temp=results.loc[results.dep==dep]
    x=np.arange(len(temp['count']))
    poly = np.polyfit(x,temp['count'],7)
    poly_y = np.poly1d(poly)(x)
#     plt.plot (temp.date, poly_y)
    plt.plot (temp.date, temp['count'])

    
plt.legend(labels=results.dep.unique())
plt.show()
```

```python
data=maindata.loc[maindata.date>'2012-01-01']

results=pd.DataFrame()

for dep in data.dep.unique():
    temp=data.loc[data.dep==dep]
    aa=temp.groupby(pd.Grouper(freq='M', key='date'))['duration'].mean().reset_index()
    aa.columns=['date','count']
    aa['dep']=dep
    results=results.append(aa)
plt.figure(figsize=(10,10))


deps=['פנימית ט', 'פנימית א', 'פנימית ב', 'פנימית ג', 'פנימית ד',
       'פנימית ה', 'פנימית ו']
rel_deps=['פנימית א','פנימית ה', 'פנימית ו']

for dep in rel_deps:
    temp=results.loc[results.dep==dep]
    x=np.arange(len(temp['count']))
    poly = np.polyfit(x,temp['count'],1)
    poly_y = np.poly1d(poly)(x)
#     plt.plot (temp.date, poly_y)

    plt.plot (temp.date, temp['count'])

plt.legend(labels=rel_deps)
plt.show()
```
